```python
import numpy as np
import pandas as pd
```

```python
data=pd.read_csv("/content/animals.csv", header='infer')
```

```python
data['Class'] = data['Class'].replace([2, 3, 4, 5, 6, 7], 0)
```

```python
# Split the data into features and target variable
X = data.drop(columns=['Sr', 'Class'])
y = data['Class']
```

```python
# Split the dataset into training and testing sets
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn import tree
```

```python
y=data['Class']
```

```python
X=data.drop(['Sr','Class'],axis=1)
```

```python
clf=tree.DecisionTreeClassifier(criterion='entropy', max_depth=3)
```

```python
clf=clf.fit(X,y)
```

```python
clf
```

```python
import pydotplus
```

```python
from IPython.display import Image
```

```python
dot_data=tree.export_graphviz(clf, out_file=None, feature_names=X.columns,
                              filled=True)
```

```python
graph=pydotplus.graph_from_dot_data(dot_data)
```

```python
# Predict on the test set
y_pred_test = clf.predict(X_test)
```

```python
# Create a DataFrame to compare actual and predicted values
comparison_df = pd.DataFrame({
    'Actual': y_test,
    'Predicted': y_pred_test
})
```

```python
# Calculate the accuracy of the model on the test set
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, y_pred_test)
accuracy
```
